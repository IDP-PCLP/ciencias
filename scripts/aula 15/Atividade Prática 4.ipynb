# Agrupamento (Clustering)

Agrupamento ou Clustering é uma técnica de aprendizado de máquina em que o conjunto de dados completo é dividido em "grupos de entidades similares". É útil para organizar conjuntos de dados grandes em grupos, ou clusters, menores que podem ser úteis para que possamos tomar ações. Por exemplo, sua base de consumidores pode ter  mais de 1 milhão de entradas, podemos tentar agrupar consumidores de alto valor, baixo valor e assim por diante.

## Que tipo de questões agrupamento tipicamente responde?

* Quais tipos de páginas existem na Web?
* De quais tipos de consumidores consistem meus clientes?
* Que tipos de pessoas estão em uma rede social?
* Que tipos de e-mail estão na minha caixa de entrada?
* Que tipos de genes o genoma humano possui?

## De agrupamento à classificação

Agrupamento pode ser o pontapé inicial para a classificação. Podemos começar agrupando nossas instâncias para depois determinar classes para cada um dos grupos.  

## Tipos de algoritmos de agrupamento

### Partitional Clustering

Agrupamento por particionamento é quando o conjunto é dividido em grupos que não se sobrepõem. 
* Funciona bem  para formas esféricas
* Eficiente e escalável
* Não é adequado para formas complexas
* Não é adequado para diferentes densidades

### Hierarchical Clustering

Abordagem determinística, não é probabilística. Divide o conjunto de acordo com uma estrutura hierárquica. Pode ser aglomerativo ou divisivo. 
* Funciona bem para relacionamentos
* Computacionalmente caro
* Sensível a ruído e outliers

### Spectral Clustering

Utiliza a representação em grafos dos dados em que cada nó representa uma entidade e o peso de cada lado. Para separar os grupos, elimina os lados de menor peso.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import SpectralClustering


from sklearn.datasets import make_moons

X3, y3 = make_moons(250, noise=.075, random_state=22)
```

```python
plt.scatter(X3[:, 0], X3[:, 1], s=50, cmap='cividis')
```

```python
# Partitional Clustering - KMeans

labels = KMeans(2, random_state=0).fit_predict(X3)
plt.scatter(X3[:, 0], X3[:, 1], c=labels, s=50, cmap='cividis')
```

```python
# Hierarchical Clustering - AgglomerativeClustering
"""
The key hyperparameter in the agglometarive clustering is the called **linkage**. It is the distance between two clusters in general. It is similar to the cluster mean M that is taken for the k-Means clustering. It can be represented in many ways:
1. Single linkage: The distance between two closest points between the two clusters
2. Complete linkage: Distance between two farthest points between the two clusters
3. Average linkage: It is between single and complete linkage. Average between all pair of points is taken. This is robust to noise
"""
single = AgglomerativeClustering(n_clusters=2, linkage='single')

labels = single.fit_predict(X3)
plt.scatter(X3[:, 0], X3[:, 1], c=labels, s=50, cmap='cividis')
```

```python
# Spectral Clustering

single = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', assign_labels='kmeans')

labels = single.fit_predict(X3)
plt.scatter(X3[:, 0], X3[:, 1], c=labels, s=50, cmap='cividis')
```
