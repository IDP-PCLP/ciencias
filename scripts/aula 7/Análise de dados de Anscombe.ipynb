## O quarteto de Anscombe

Por que a visualização de dados é necessária quando temos ferramentas estatísticas para lidar com o conjunto de dados? 

Considere os quatro conjuntos de dados abaixo. Todos estão no mesmo DataFrame, separados pela coluna 'dataset'.

```python
import seaborn as sns
sns.set_theme(style="ticks")

anscombe = sns.load_dataset("anscombe")
descricao = anscombe.groupby('dataset').describe()
print(descricao)
```

Percebam que cada conjunto de dados possui um comportamento completamente distinto, porém todos os indicadores da estatística descritiva são idênticos. 

```python
anscombe.groupby('dataset').plot.scatter(x='x',y='y',grid=True,color='blue')
```

```python
sns.scatterplot(x='x',y='y',data=anscombe,hue='dataset')
```

```python
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=anscombe,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```
